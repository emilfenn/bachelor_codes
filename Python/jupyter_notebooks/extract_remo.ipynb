```python
import xarray as xr 
import pandas as pd 
from glob import glob 
import os 

base_path = "/work/ch0636/g300147/remo-results/2019" 

# Coordinates of the station 
lat0 = -26.150 
lon0 = 28.00 

files = sorted(glob(os.path.join(base_path, "e067003e_c0167_2019*.nc"))) 

for path in files: 
    print("Working on:", os.path.basename(path)) 
    ds = xr.open_dataset(path) 
    pixel = ds.sel(rlat=lat0, rlon=lon0, method="nearest") 
    temp_series = pixel["TEMP2"].isel(height2m=0) 
    
    df = temp_series.to_dataframe().reset_index() 
    
    # Monat aus Dateiname holen (…_201612.nc → 201612) 
    
    yyyymm = os.path.basename(path).split("_")[-1].replace(".nc", "") 
    
    outname = f"temp2_{yyyymm}_Roosevelt_Park.csv" 
    
    #Name of station 
    df.to_csv(outname, index=False) 
    ds.close()
```
